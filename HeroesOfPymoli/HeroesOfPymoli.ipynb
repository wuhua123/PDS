```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Create data frames to hold data

male_players = purchase_data.loc[purchase_data['Gender'] == "Male",:]
female_players = purchase_data.loc[purchase_data['Gender'] == "Female",:]
others = purchase_data.loc[(purchase_data['Gender'] != "Male") & (purchase_data['Gender'] != "Female"),:]

# Calculate values

count_males = len(male_players['SN'].value_counts())
count_females = len(female_players['SN'].value_counts())
count_others = len(others['SN'].value_counts())

total_players = count_males+count_females+count_others

# Formatting

percent_male = "{:,.2f}".format((count_males/total_players)*100)
percent_female = "{:,.2f}".format((count_females/total_players)*100)
percent_others = "{:,.2f}".format((count_others/total_players)*100)

gender_demo = pd.DataFrame({"Total Count": [count_males, count_females, count_others],
                           "Percentage of Players": [percent_male, percent_female, percent_others],
                           "": ["Male", "Female", "Other / Non-Disclosed"]})
# Adjust apperance
gender_demo.set_index("", inplace=True)

# Print
gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Caculate values

pur_count_m = male_players["Purchase ID"].count()
pur_count_f = female_players["Purchase ID"].count()
pur_count_o = others["Purchase ID"].count()

total_value_m = male_players["Price"].sum()
total_value_f = female_players["Price"].sum()
total_value_o = others["Price"].sum()

ave_price_m = "${:,.2f}".format(total_value_m/pur_count_m)
ave_price_f = "${:,.2f}".format(total_value_f/pur_count_f)
ave_price_o = "${:,.2f}".format(total_value_o/pur_count_o)

person_count_m = len(male_players["SN"].value_counts())
person_count_f = len(female_players["SN"].value_counts())
person_count_o = len(others["SN"].value_counts())

ave_value_m = "${:,.2f}".format(total_value_m/person_count_m)
ave_value_f = "${:,.2f}".format(total_value_f/person_count_f)
ave_value_o = "${:,.2f}".format(total_value_o/person_count_o)

# Create data frame with values

Pur_Analysis = pd.DataFrame({"Purchase Count": [pur_count_f, pur_count_m, pur_count_o],
                            "Average Purchase Price": [ave_price_f, ave_price_m, ave_price_o],
                            "Total Purchase Value": [total_value_f, total_value_m, total_value_o],
                            "Avg Total Purchase per Person": [ave_value_f, ave_value_m, ave_value_o],
                            "Gender":["Female", "Male", "Other / Non-Disclosed"]})
# Formatting
Pur_Analysis["Total Purchase Value"] = pd.to_numeric(Pur_Analysis["Total Purchase Value"], errors='coerce').map("${:,}".format)

# Adjust appearance
Pur_Analysis.set_index('Gender', inplace=True)

# Print
Pur_Analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Make a copy of DataFrame - purchase_data, use the copy to produce data set with unique player in each row
purchase_data_copy = purchase_data.copy(deep=True)
unique_player_data = purchase_data_copy.drop_duplicates(subset="SN", keep="first", inplace=False)

# Set bins and label names
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Slice data into bins according to age
unique_player_data["Categorized by Age"] = pd.cut(unique_player_data["Age"], bins, labels=group_names)

# Calculate total number of players
total_players = unique_player_data['SN'].count()

# Calculate total players by age category, players percent by age
players_grouped_byAge = unique_player_data.groupby(["Categorized by Age"])
total_players_byAge = players_grouped_byAge["SN"].count()
percent_byAge = (total_players_byAge/total_players)*100

# Place value in data frame
Age_Demo = pd.DataFrame({"Total Count": total_players_byAge,
                         "Percentage of Players": percent_byAge})
# Format value in data frame
Age_Demo["Percentage of Players"] = Age_Demo["Percentage of Players"].map("{:,.2f}".format)

# Adjust appearance
Age_Demo.reset_index(inplace=True)
Age_Demo = Age_Demo.rename(columns={"Categorized by Age":""})
Age_Demo.set_index("", inplace=True, )

# print data frame
Age_Demo
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group by spender
grouped_purchase_data = purchase_data.groupby('SN')

# Basic calculations
purchase_count = grouped_purchase_data['Purchase ID'].count()

total_purchase_value = grouped_purchase_data['Price'].sum()  

ave_purchase_price = total_purchase_value/purchase_count\

# Create data frame to hold data
spenders_data = pd.DataFrame({"Purchase Count": purchase_count,
                             "Average Purchase Price": ave_purchase_price,
                             "Total Purchase Value": total_purchase_value})

# Sort table according to total purchase value to find toppest spenders
spenders_data = spenders_data.sort_values(['Total Purchase Value'], ascending=False)

# Format value
spenders_data["Average Purchase Price"] = spenders_data["Average Purchase Price"].map("${:,.2f}".format)

spenders_data["Total Purchase Value"] = spenders_data["Total Purchase Value"].map("${:,.2f}".format)

# print toppest 5 spenders
spenders_data.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name and Price columns
items = purchase_data[["Item ID", "Item Name", "Price"]]

# Grouped by Item ID and Item Name
grouped_items = items.groupby(['Item ID', 'Item Name'])

# Calculate purchase count, item price
purchase_count = grouped_items['Price'].count()
purchase_count = pd.DataFrame({"Purchase Count": grouped_items['Price'].count()})
purchase_count.reset_index(inplace=True)

unique_items = items.drop_duplicates(subset='Item ID')
item_price = unique_items.loc[:, ['Item ID', 'Item Name', 'Price']]

# Create data frame to hold results
popular_items = pd.merge(purchase_count, item_price, on=["Item ID", "Item Name"])

# Put total purchase value into data frame
popular_items["Total Purchase Value"] = popular_items['Purchase Count']*popular_items['Price']

# Sort by purchase count
popular_items = popular_items.sort_values(['Purchase Count'], ascending=False)

# Adjust appearance
popular_items.set_index(["Item ID", "Item Name"], inplace=True)

# Print
pd.options.display.float_format = '${:,.2f}'.format
popular_items.head()
```
